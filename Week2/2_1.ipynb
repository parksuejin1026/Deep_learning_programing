# 2мһҘ. л”Ҙлҹ¬лӢқ н•өмӢ¬ лҜёлҰ¬ліҙкё°

лӮҳмқҳ мІ« л”Ҙлҹ¬лӢқ : '10мһҘ нҸҗм•” мҲҳмҲ  нҷҳмһҗмқҳ мғқмЎҙмңЁ мҳҲмёЎ' мҪ”л“ң лҜёлҰ¬ліҙкё°

```python
from tensorflow.keras.models import Sequential # н…җм„ңн”ҢлЎңмқҳ мјҖлқјмҠӨ APIм—җм„ң н•„мҡ”н•ң н•ЁмҲҳ л¶Ҳлҹ¬мҳӨкё°
from tensorflow.keras.layers import Dense # лҚ°мқҙн„°лҘј лӢӨлЈЁлҠ” лҚ° н•„мҡ”н•ң лқјмқҙлёҢлҹ¬лҰ¬
import numpy as np # л„ҳнҢҢмқҙ
```

```python
!git clone https://github.com/taehojo/data.git # к№ғн—ҲлёҢ нҒҙлЎ 

Data_set = np.loadtxt("./data/ThoraricSurgery3.csv", delimiter= ',') # нҷҳмһҗ лҚ°мқҙн„°
X = Data_set[:,0:16] # нҷҳмһҗмқҳ м§„м°° кё°лЎқ X м§Җм •
y = Data_set[:,16] # нҷҳмһҗмқҳ мӮ¬л§қ/мғқмЎҙ м—¬л¶Җ y
```

```python
model = Sequential() # л”Ҙлҹ¬лӢқ лӘЁлҚёмқҳ кө¬мЎ° Sequential
model.add(Dense(30, input_dim=16, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
history = model.fit(X, y, epochs = 500, batch_size = 16)
```
